```python
#import module
from google_play_scraper import app
import pandas as pd
import numpy as np
```

```python
#scrapping ulasan sebanyak 1000
from google_play_scraper import Sort,reviews

result,continuation_token=reviews(
    'com.telkom.tracencare',
    lang='id', #bahasa yang digunakan=indonesia
    country= 'id',
    sort=Sort.NEWEST, #memilih data yang terbaru
    count=1000, #data yang diambil 1000
    filter_score_with=None #jika ingin mengambil ulasan sesuai rating 
)
```

```python
df_busu = pd.DataFrame(np.array(result),columns=['review'])
df_busu = df_busu.join(pd.DataFrame(df_busu.pop('review').tolist()))
df_busu.head()
```
